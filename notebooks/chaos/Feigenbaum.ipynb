# Logistic Map
What happens to a simple model of population growth when exploring different parameters of the model. We create a simple equation to serve as the model. Given a fixed fertility $\lambda$ and a value for births $B_n$ and $D_n$ per fixed time period $n$, we can model the next population time period $n+1$ as simple function
$P=\lambda B_n D_n$. We can further simplify this equation and our model by considering only births and deaths as simply $(1-B_n)$. Thus, renaming $B_n$ to $x_n$ to match standard notation, our equation becomes an iterative function $x_{n+1}=\lambda x_n (1-x_n)$.

We begin with a simple population $x_n=0.5$ and set the fertility to a fixed value $\lambda=2.3$. We observe the population over time and see what occurs

```python
N = 30 # number of steps to compute
x = 0.5
Lambda = 2.3
```

```python
import numpy as np

T = np.arange(N) # time axis values
X = np.zeros(N) # track population over time
for index, t in enumerate(T):
    x = Lambda*x*(1-x)
    X[index] = x
```

```python
#plot line
import matplotlib.pyplot as plt

plt.plot(T,X, 'r') #plot as red line
plt.show()
```

The result is a a steady state where the population has stablised.

Now we have a steady state involving four stable population values and transitioning between them over time. Now lets explore a range of all possible $\lambda$ values and look at the resulting plot to see how many steady states there are over different fertility values.

```python
#for convenience define logistic function given parameters
def logistic(r, x):
    return r * x * (1 - x)
```

```python
N = 1000
last = 100
deltaLambda = 1000 #resolution of the lambda search
Lambda = np.linspace(2.5, 3.5, deltaLambda)
x = 1e-5 * np.ones(deltaLambda) #initial point the same for all r

fig, ax = plt.subplots(1, 1, figsize=(16, 10))

for i in range(N):
    x = logistic(Lambda, x)

    #Display the plot
    if i >= (N - last):
        ax.plot(Lambda, x, ',k', alpha=.25)
plt.show()
```
